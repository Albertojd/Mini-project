```python
import words
import random
import string
failure = ['''
      +---+
      |   |
          |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========
    YOU LOST''']
word_list = ["python", 'ironhack', 'dog', 'cat', 'mint', 'earth', 'bottle', 'book', 'android', 'pen',
         'computer', 'glasses', 'cork', 'bicycle', 'guitar', 'screwdriver',"ball", "headphones", "screen", "windows", "usb"]

chosen_word = random.choice(word_list) ##Elige una palbra aletoriamente
errores = 0  ### Varaible para contabilizar errores
progreso = []  ### Lista que se rellenará con espacios en blancos, para luego ir incluyendo las letras acertadas
cantidad_letras= len(chosen_word) ### Esto nos ayuda a pintar el número de letras que tiene la palabra oculta
alphabet= list(string.ascii_lowercase)

for i in range(len(chosen_word)): ### Este bucle rellena la lista de progreso con los espacios en blanco correspondientes
    progreso.append("_ ")

check_word = [] ### Nos ayudará a comprobar si coincide la palabra escrita con la elegida por el programa
for char in chosen_word:
    check_word.append(char+" ")
letras_usadas=[]

while errores < 7: ### Mantiene el juego activo hasta que tengamos 7 fallos, el programa pinta al ahorcado
    if errores == 0:
        print(failure[0])
    elif errores == 1:
        print(failure[1])
    elif errores == 2:
        print(failure[2])
    elif errores == 3:
        print(failure[3])
    elif errores == 4:
        print(failure[4])
    elif errores == 5:
        print(failure[5])
    elif errores == 6:
        print(failure[6])
        break
    
    
    print("".join(progreso)) ###Convertimos la lista, en un string, y la pintamos
    print("La palabra tiene: ",cantidad_letras, "letras")  ### Con esto ayudamos a ver cuantas letras tiene la palabra oculta
    print("Letras usadas: ", letras_usadas)  ### Pinta las letras que ya has usado

    letra= input("Elige una letra:")
    if letra in letras_usadas:
        print("ya usaste esta letra...")
    if letra not in alphabet:
        print("Por favor, sin letras mayusuclas, o simbolos raros")
    else:
        letras_usadas.append(letra)
        
        hay_error = True  ##Sí la letra está mal
        for i in range(len(chosen_word)):
            if letra == chosen_word[i]:
                progreso[i]= letra + " "
                hay_error = False
        
        if hay_error:
            errores += 1
    
        if check_word == progreso:
            print("".join(progreso))
            print("YOU WIN!, The word is: ", chosen_word.upper())
            break
```
